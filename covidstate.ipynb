```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas import DataFrame as df
from datetime import datetime
import numpy as np
import covidutil as cu
```

```python
#create file path

folder = "c:/PY-ML/data/covid1906/"
file = folder + "states.csv"
```

```python
#create dataframe
df = pd.read_csv(file)
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df["Date"])
df['Day'] = df['Date'].dt.day_name()
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Active'] = df['Confirmed'] - df['Recovered'] - df['Deceased']
```

```python
state_list = list(df['State'].unique())
```

```python
len(state_list)
```

```python
def get_state(data,state):
    st = data[data['State'] == state].reset_index(drop = True)
    return st
```

```python
all_state_data = []
for i in state_list:
    data = get_state(df,i)
    all_state_data.append(data[:])
```

```python
#subplot of all states
for i in range(0,10):
    plt.style.use("grayscale")
    
    fig, (ax1,ax2,ax3,ax4) = plt.subplots(1,4,figsize = (20,3),sharex = True)
    
    plt.ticklabel_format(style = ('scientific'))
    plt.ticklabel_format(style = ('scientific'))
    ax1.plot(all_state_data[i]['Confirmed'])
    ax1.set_title(state_list[i] + ' Confirmed')
    ax1.set_xlabel('days')
    ax1.set_ylabel('Confirmed')
    
    plt.ticklabel_format(style = ('scientific'))
    ax2.plot(all_state_data[i]['Deceased'])
    ax2.set_title(state_list[i] + ' Deceased')
    ax2.set_xlabel('days')
    ax2.set_ylabel('Deceased')
    
    plt.ticklabel_format(style = ('scientific'))
    ax3.plot(all_state_data[i]['Recovered'])
    ax3.set_title(state_list[i] + ' Recovered')
    ax3.set_xlabel('days')
    ax3.set_ylabel('Recovered')
    
    plt.ticklabel_format(style = ('scientific'))
    ax4.ticklabel_format(style = 'plain')
    ax4.plot(all_state_data[i]['Active'])
    ax4.set_title(state_list[i] + ' Active')
    ax4.set_xlabel('days')
    ax4.set_ylabel('Active')
```
